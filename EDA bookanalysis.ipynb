```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("books_1000.csv")

df.head()
```

```python
print("Rows and Columns:", df.shape)

df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.head()
```

```python
df["Price"] = (
    df["Price"]
    .astype(str)
    .str.extract(r'(\d+\.\d+)')[0]
    .astype(float)
)
```

```python
print(df["Price"].dtype)
```

```python
print("Average Price:", round(df["Price"].mean(), 2))
```

```python
print("Highest Price:", df["Price"].max())
print("Lowest Price:", df["Price"].min())
```

```python
top10 = df.sort_values(
    by="Price",
    ascending=False
).head(10)

top10
```

```python
cheap10 = df.sort_values(
    by="Price",
    ascending=True
).head(10)

cheap10
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Price"], bins=20)

plt.title("Distribution of Book Prices")
plt.xlabel("Price")
plt.ylabel("Number of Books")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.boxplot(df["Price"])

plt.title("Book Price Box Plot")

plt.show()
```
